# Echo Chambers
## Basic Considerations

This Notebook models the time evolution of opinions in an agent-based system, based on [_Baumann et al, Modeling echo chambers and polarization dynamics in social networks_](https://arxiv.org/pdf/1906.12325.pdf).

The simple model assumes N agents, each with a one-dimensional opinion and a one-dimensional activiy, based on which an agent is more or less likely to share its opinion. The network between these agents is freshly initialized for each time step, primarily depending on homophily (i.e. similarity of opinions) between two agents.

Next to reproduction of _Baumann et al_'s Figure 1 (see below), this project is concerne with the impact of so-called _influencers_, i.e. additional agents more active and communicative than common agents.

```python
# Standard mathematics and plotting libraries
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# For random sampling during network matrix generation
from numpy.random import default_rng

# For diff. eq. solving via Runge-Kutta 4
from scipy.integrate import odeint

# For activities random sampling according to own power law probability density
from scipy.stats import rv_continuous

# For execution time measurements
import time

# For progress bar
from tqdm import tqdm
```

## The Model
In the following cell the class agents and model is defined. Agents contain their opinion x and their activity a, with their respective getter and setter functions. The class model can be initiated with either the number of agents and their activity distribution, the Baumann model, or additionaly with a number of influencer and an array of activity for the influencer. The opinion of the influencer are set between [-2.5;2.5] to represent the opinion space in Fig. 1. Finally, the distribution class for the power law is defined.

```python
# Define classes
class Agent:
    def __init__(self, opinion, activity):
        self.x=opinion
        self.a=activity
        
    def setOpinion(self, opinion):
        self.x=opinion   
        
    def getOpinion(self):
        return self.x
    
    def setActivity(self, activity):
        self.a=activity
        
    def getActivity(self):
        return self.a
        
class Model:
    def __init__(self, N, a_dist):
        
        x = np.linspace(-1,1,N)
        PowerDistribution = PowerLaw(a=epsilon)
        a = a_dist.rvs(size=N)
        self.network = []
        for i in range(N):
            self.network.append(Agent(x[i], a[i]))
    
    def __init__(self, N, N_Influencer, a_dist, a_Influencer):
        
        # Initialise N common agents
        x = np.linspace(-1,1,N)
        a = a_dist.rvs(size=N)
        self.network = []
        for i in range(N):
            self.network.append(Agent(x[i], a[i]))
        
        # Initialise N_Influencer influencer agents
        x = np.linspace(-2.5,2.5,N_Influencer)
        self.influencers = []
        for i in range(N_Influencer):
            self.influencers.append(Agent(x[i], a_Influencer[i]))
        
    #Define Getters and Setters
    
    def getNetwork(self):
        return self.network
    
    def getInfluencers(self):
        return self.influencers
    
    def getAgentsOpinions(self):
        allOpinions = np.zeros(len(self.network))
        for i in range(len(self.network)):
            allOpinions[i] = self.network[i].getOpinion()
        return allOpinions
    
    def getInfluencerOpinions(self):
        influencerOpinions = np.zeros(len(self.influencers))
        for i in range(len(self.influencers)):
            influencerOpinions[i] = self.influencers[i].getOpinion()
        return influencerOpinions
    
    def getAgentsActivities(self):
        allActivities = np.zeros(len(self.network))
        for i in range(len(self.network)):
            allActivities[i] = self.network[i].getActivity()
        return allActivities
    
    def getInfluencerActivities(self):
        influencerActivities = np.zeros(len(self.influencers))
        for i in range(len(self.influencers)):
            influencerActivities[i] = self.influencers[i].getActivity()
        return influencerActivities
    
    def setAgentsOpinions(self, newOpinions):
        if len(self.network) == len(newOpinions):
            for i in range(len(self.network)):
                self.network[i].setOpinion(newOpinions[i])
    
    def setInfluencerOpinions(self, newOpinions):
        if len(self.influencers) == len(newOpinions):
            for i in range(len(self.influencers)):
                self.influencers[i].setOpinion(newOpinions[i])
    

class PowerLaw(rv_continuous):
    """Activitiy sampling distribution identical to Baumann et al.
    It models the intuitive assumption, that by far most people
    are hardly posting on social media, and even less people being
    very active. 
    

    For Reference, see https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.rv_continuous.html#scipy.stats.rv_continuous"""
    # define probability distribution as the paper's "F(a)"
    def _pdf(self, x):
        return ((1-gamma)/(1-epsilon**(1-gamma)))*x**(-1*gamma)
```

## ODE and contact matrix
The used ODE is defined here. This is followed by the functions to calculate the contact matrix A for the model with and without influencer. 

```python
# Define methods
from timeit import default_timer

# Implement differential eq. 'dx/dt = [...]'
def diffEq(x, t, alpha, K):
    N=len(x)
    sol = np.zeros(N)
    for index_i, i in enumerate(x):
        sum_tmp = 0
        for index_j, j in enumerate(x):
            sum_tmp +=A[index_i][index_j]*np.tanh(alpha*j)
        sol[index_i] = -i+K*sum_tmp
    return sol

#method that calculates the probabilites an agent gets activated
def calculateProbabilities(i, j, x, beta):
    sum_tmp = 0
    for j_tmp in range(len(x)):
        if x[i] != x[j_tmp]:
            sum_tmp = sum_tmp+abs(x[i]-x[j_tmp])**(-beta)
    return (abs(x[i]-x[j])**(-beta))/sum_tmp
    
#method that generates the contact matrix for the base model
def generateContactMatrix(model, N, beta, m, r):
    x = np.array(model.getAgentsOpinions())
    a = np.array(model.getAgentsActivities())
    rng = default_rng()
    A = np.zeros((len(x), len(x)))
    
    for index_i, i in enumerate(x):

        if a[index_i] < rng.random(): continue
        sum_tmp=0
        for j in x:
            if i==j: continue
            sum_tmp +=abs(i-j)**(-beta)
        
        for index_j, j in enumerate(x):
            if i==j: continue
            A[index_i][index_j] = (abs(i-j)**(-beta))/sum_tmp
            if m*A[index_i][index_j] > rng.random():
                A[index_i][index_j] = 1
                # Füge recipr. contacting ein:
                if r > rng.random():
                    A[index_j][index_i] = 1
    return A

#method that generates thes contact matrix including influencers for the extended model
def generateContactMatrix_extended(model, N, beta, m,m_Influencer, r):
    
    x = np.array(model.getAgentsOpinions())
    x = np.append(x, model.getInfluencerOpinions())
    a = np.array(model.getAgentsActivities())
    a = np.append(a, model.getInfluencerActivities())
    
    rng = default_rng()
    A = np.zeros((len(x), len(x)))
    for index_i, i in enumerate(x):
        if a[index_i] < rng.random(): continue
        
        sum_tmp=0
        for j in x:
            if i==j: continue
                # calculate sum used in calculation of contacting probability p_ij
            sum_tmp +=abs(i-j)**(-beta)
        
        for index_j, j in enumerate(x):
            # index to go through different influencer audiences, given in array m_influencer
            c = 0
            
            if i==j or A[index_j][index_i]==1: continue
            A[index_j][index_i] = (abs(i-j)**(-beta))/sum_tmp
            if index_i < N and m*A[index_j][index_i] > rng.random():
                A[index_j][index_i] = 1
                if r > rng.random():
                    A[index_j][index_i] = 1
            elif m_Influencer[c]*A[index_j][index_i] > rng.random():
                c = c+1
                A[index_j][index_i] = 1
                # Füge recipr. contacting ein:
                if r > rng.random():
                    A[index_i][index_j] = 1
    return A

#method for plotting 
def plotGraph(data, ax):
    times = np.arange(0, len(data)*dt, dt)
    
    for i in range(len(data.T)):
    
        # In case the last value of agent i is negative, the graph color will be changed to red 
        if data.T[i][-1] < 0:
            color = 'r'
            LW=.25
        # Else if the last value i is positive, the graph color will be blue
        else:
            color ='b'
            LW=.25
        # Influencers will be coloured black
        if i>=N: 
            #opinionStorage.T[i]= x_influencer[i-N]
            color = 'black'
            LW=1
        ax.plot(times, data.T[i], color, linewidth=LW)
```


## Model Parameter 

In this section the parameter of the model are defined and the model itself is created. Notably, the variable m_Influencer is the percentage of the network one single influencer can reach.

```python
# Define fixed parameters, FOR DIFFERENT RESULTS CHANGE PARAMETERS HERE
t_current = 0
tMax = 10
dt = 0.01
N = 100
N_Influencer = 7
K=3
# Define different alpha, beta values as arrays from which to pick for each graph
alphas = np.array([0.05, 3, 3])
betas = np.array([2, 0, 3])
#alpha_current = alphas[1]
#beta_current = betas[1]

# Fixed activity-driving (AD) parameters
m = 10
m_inf=0.2
m_Influencer = np.array((N_Influencer)*[m_inf*N])
a_inf=0.8
a_Influencer = np.full(N_Influencer, a_inf)
epsilon = 0.01
gamma = 2.1
r = 0.5
a_dist=PowerLaw(a=epsilon)
```

## Base Model
In this section we set up the base model by processing it first and then plotting the results

### Processing

```python
for i in range(len(alphas)):
    model = Model(N, N_Influencer, a_dist, a_Influencer)
    # the current time step
    t_current = 0
    # the current alpha (controversialness) value set in the parameters section
    alpha_current = alphas[i]
    # the current beta (homophily) value set in the parameters section
    beta_current = betas[i]

    print('Computing Model', i+1, '/', len(alphas))
    
    opinionStorage = np.array(model.getAgentsOpinions())
    
    # Integrate dynamics with 4th order Runge-Kutta (RK4)
    for t in tqdm(np.arange(0, tMax, dt)):
        t_current = t_current + dt
        
        t_RK4 = np.linspace(t_current-dt, t_current, 2)

        currentOpinions = np.array(model.getAgentsOpinions())
        currentActivities = np.array(model.getAgentsActivities())
        A = generateContactMatrix(model, N, beta_current, m, r)
        
        #currentOpinions = np.append(currentOpinions, model.getInfluencerOpinions())
        solODE = odeint(diffEq, currentOpinions, t_RK4, args=(alpha_current, K))
        
        # Update and store new opinions in agents and storage array
        newOpinions = solODE[-1]
        model.setAgentsOpinions(newOpinions)       
        opinionStorage = np.vstack([opinionStorage, newOpinions])
        
        # store the opinions for the 3 alpha/beta value combinations into 3 different arrays for later plotting
        match i:
            case 0:
                resultArray1 = np.array(opinionStorage)
            case 1:
                resultArray2 = np.array(opinionStorage)
            case 2:
                resultArray3 = np.array(opinionStorage)
```

### Discussion of the results
With the introduction of homophily and a small value of alpha, the controversialness, set at .05 the opinions converge towards a neutral consensus.
In absence of homophily, if beta equals 0, and for a higher value of alpha set at 3, the system becomes radicalized. In our case, you can see how this radicalization towards one of both sides, as our model converged towards the opposite extreme of the spectrum
If beta is reintroduced, the system becomes a polarized state.


## Model Extension
In this section we extend the model. Up until now, two values had been static. One was the parameter m, which stands for the amount of peers an active agent will contact. The other is r, the reciprocity parameter, or in other words, the probability that if an agent has an interaction with a peer, that this peer will also interact back to said agent.  On the other hand, the activity a had been defined as a probability for an agent to be active in any given period. If active, an agent would contact m peers and influence them with another probability which depended the homophily beta.
We define an Influencer as an agent that has a high value of m and a high fixed value for the activity a. 
Their opinion will be fixed, meaning that they won’t change it, nor can other agents influence them, as their x value won’t be updated with the Equation presented earlier. Therefore, their r–value isn’t relevant and equals indirectly 0. Their x-value is equally distributed across [-2.5;2.5].  
We hypothesised that the opinions of other Agents will move in the direction in which an influencer’s opinion is located in the opinion-spectrum xi in [-2.5;2.5] and the state of a Regime will depend on where these influencers have situated their opinions.

### Processing

```python
for i in range(len(alphas)):
    model = Model(N, N_Influencer, a_dist, a_Influencer)
    # the current time step
    t_current = 0
    # the current alpha (controversialness) value set in the parameters section
    alpha_current = alphas[i]
    # the current beta (homophily) value set in the parameters section
    beta_current = betas[i]
    #xecution_start_time = time.time()
    print('Computing Model', i+1, '/', len(alphas))
    opinionStorage = np.array(model.getAgentsOpinions())   
    opinionStorage = np.append(opinionStorage, model.getInfluencerOpinions())


    # Integrate dynamics with 4th order Runge-Kutta (RK4)

    for t in tqdm(np.arange(0, tMax, dt)):
        t_current = t_current + dt

        t_RK4 = np.linspace(t_current-dt, t_current, 2)
        A = generateContactMatrix_extended(model, N, beta_current, m, m_Influencer, r)
        currentOpinions = np.array(model.getAgentsOpinions())
        currentOpinions = np.append(currentOpinions, model.getInfluencerOpinions())
        solODE = odeint(diffEq, currentOpinions, t_RK4, args=(alpha_current, K))

        # Update and store new opinions in agents and storage array
        newOpinions = solODE[-1]
        model.setAgentsOpinions(newOpinions[:N])
        #model.updateInfluencerOpinions(newOpinions[N:])

        newOpinions = np.array(model.getAgentsOpinions())
        newOpinions = np.append(newOpinions, model.getInfluencerOpinions())

        opinionStorage = np.vstack([opinionStorage, newOpinions])
        
        # store the opinions for the 3 alpha/beta value combinations into 3 different arrays for later plotting
        match i:
            case 0:
                ext_resultArray1 = np.array(opinionStorage)
            case 1:
                ext_resultArray2 = np.array(opinionStorage)
            case 2:
                ext_resultArray3 = np.array(opinionStorage)

```

### Disussion of the results
On the neutral consensus regime you can see how it takes longer for opinions move into that static state and that there is not a clear convertion anymore. You can rather see a bigger interval of opinions shared around x = 0.

In the polarization regime you can see how the spectrum around the neutral consensus tends to be wider, where as the polarized opinions present a much smaller window of how the x value may vary. We called these opinion clusters. These are small intervals of an opinion on the spectrum with high amounts of agents sharing that opinion

The last example shows what happens if you play with the amount of influencers in a polarized state and take away one of the neutral consensus (so alpha, beta = 3).
Here you can clearly see how an influencer at x = 0 determines if there will be any other agents with their opinion set around this value.

With a higher amount of influencers the opinion clusters move more towards the extremes. Another interesting finding is that almost no agents sets their opinion back in the direction of the neutral consensus.


## Limitations and Conclusion
We found that the outcome of a regime couldn’t be influenced in terms of polarization, radicalization or neutral cocensus, which still happened depending on the alpha and betha parameters respectively. However, the characteristics of those regimes did change, most significantly in the case of the polarized regime.

There, you could see how Influencers highly determined where the polarized opinions would cluster. This was in between two Influencers located at the respective extremes. The presence of opinions around the neutral consensus there would only be the case if an Influencer had his opinion at x = 0. If this was the case, then the window of opinions around the neutral consensus would be rather big, or in other words, here you could see a bigger interval of the amount of opinions agents had.

An Interesting question that reaches the limits of our model’s expansion is how the formation of these opinion clusters might influence the formation of social bubbles. For that, a statistical analysis of the temporal adjacency matrix would have to be done in order to see which connection each agent has and if these are just with like-minded individuals.
